```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
```

## creating dataset

```python
from sklearn.datasets import make_classification
```

```python
# create dataset 
X,y = make_classification(n_samples = 1000, n_features = 10, n_classes = 2, random_state = 42)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred = logistic.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

## Hyperparameter tuning and CV


```python
model = LogisticRegression()
penalty = ["l1", "l2", "elasticnet"]
c_values = [100, 10, 1.0, 0.1, 0.01]
solver = ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]
```

```python
params = dict(penalty = penalty, C = c_values, solver = solver)
params
```

## grid search CV

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator = model, param_grid = params, scoring = "accuracy", cv = cv, n_jobs = -1)
```

```python
print(grid)
```

```python
print(grid.fit(X_train, y_train))
```

```python
print(grid.best_params_)
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

##  Logistic Regresssion in imabalanced dataset

```python
# Creating and imbalanced dataset

from collections import Counter
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples = 10000, n_features = 2, n_clusters_per_class = 1, n_redundant = 0, weights = [0.99], random_state = 42)
```

```python
X
```

```python
Counter(y)
```

```python
pd.DataFrame(X)[0]
```

```python
pd.DataFrame(X)[1]
```

```python
import seaborn as sns
import pandas as pd

sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue = y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.25, random_state = 42)
```

```python
model = LogisticRegression()
penalty = ["l1", "l2", "elasticnet"]
c_values = [100, 10, 1.0, 0.1, 0.01]
solver = ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]

# class_weights is used in imbalanced dataset
class_weights = [{0:w, 1:y} for w in [1, 10, 50, 100] for y in [1, 10 , 50 , 100] ]
```

```python
class_weights
```

```python
params = dict(penalty = penalty, C= c_values, solver = solver, class_weight = class_weights)
```

```python
params
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
cv = StratifiedKFold()
grid = GridSearchCV(param_grid = params, estimator= model, scoring = "accuracy", cv = cv)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
y_pred
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```
