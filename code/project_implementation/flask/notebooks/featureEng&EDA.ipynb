```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("forest_fire.csv", header = 1)
df
```

```python
df.head()
```

```python
df.info()
```

## Data Cleaning

```python
# missing values

df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, "Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df
```

```python
df.info()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

## removing the null vals

```python
df = df.dropna().reset_index(drop = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

## fix spaces in columns names

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

#### changes the required columns to integer datatype

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

#### changing the other columns to float datatype



```python
objects = [features for features in df.columns if df[features].dtypes == "O"]
```

```python
for i in objects:
    if i != "Classes":
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

## save the clean dataset

```python
df.to_csv("forest_fires_cleaned.csv", index = False)
```

## Exploratory data analysis

```python
# dropping columns like date month and year

df_copy = df.drop(["day","month","year"], axis = 1)
df_copy.head()
```

```python
df_copy["Classes"].value_counts()
```

```python
# encoding of categories in classes
df_copy["Classes"] = np.where(df_copy["Classes"].astype(str).str.contains("not fire", na=False), 0, 1)
df_copy.tail()
```

```python
df_copy["Classes"].value_counts()
```

```python
# percentage of pie chart

percentage = df_copy["Classes"].value_counts(normalize = True) * 100
```

```python
# plot pie chart

classlabels = ["Fire", "Not fire"]
plt.figure(figsize = (12,7))
plt.pie(percentage, labels = classlabels)
plt.show()
```

```python
# correlation

df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
plt.show()
```
