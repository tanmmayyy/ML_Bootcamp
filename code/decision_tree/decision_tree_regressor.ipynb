```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
df = load_diabetes()
```

```python
print(df["DESCR"])
```

```python
df_diabetes = pd.DataFrame(df.data, columns = ["age", "sex" ,"bmi", "bp", "s1", "s2", "s3", "s4", "s5", "s6"] )
df_diabetes.head()
```

```python
X = df_diabetes
y = df["target"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(X_train.corr(), annot = True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

## Hyperparameter tuning

```python
param = {
    "criterion" : ['squared_error', 'friedman_mse', 'absolute_error'],
    "splitter" : ['best', 'random'],
    "max_depth" : [1,2,3,4,5, 10, 15 , 20 , 25],
    "max_features" : ['sqrt', 'log2', 'auto']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
grid = GridSearchCV(regressor, param_grid = param, cv = 5, scoring = "neg_mean_squared_error")
```

```python
import warnings 
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score , mean_absolute_error, mean_squared_error

print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

```python
select_model =  DecisionTreeRegressor(criterion = "squared_error", max_depth = 3, max_features = "log2", splitter = "best")
```

```python
select_model.fit(X_train, y_train)
```

```python
# visualize the decison tree

from sklearn import tree

plt.figure(figsize=(15, 10))
tree.plot_tree(select_model, filled=True)
plt.show()
```
