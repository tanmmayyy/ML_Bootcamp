```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
import seaborn as sns
```

```python
# creating synthetic datapoints

from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples = 1000, n_features = 2, n_classes = 2, n_clusters_per_class = 2, n_redundant = 0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.svm import SVC

model = SVC(kernel = "linear")
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix

cr = classification_report(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print(cr)
print(cm)
```

```python
# using different kernels

rbf = SVC(kernel = "rbf")
rbf.fit(X_train, y_train)
y_pred1 = rbf.predict(X_test)
cr = classification_report(y_test, y_pred1)
cm = confusion_matrix(y_test, y_pred1)
print(cr)
print(cm)
```

```python
# using different kernels

poly = SVC(kernel = "poly")
poly.fit(X_train, y_train)
y_pred2 = poly.predict(X_test)
cr = classification_report(y_test, y_pred2)
cm = confusion_matrix(y_test, y_pred2)
print(cr)
print(cm)
```

```python
# using different kernels

sigmoid = SVC(kernel = "sigmoid")
sigmoid.fit(X_train, y_train)
y_pred3 = sigmoid.predict(X_test)
cr = classification_report(y_test, y_pred3)
cm = confusion_matrix(y_test, y_pred3)
print(cr)
print(cm)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C' : [0.1, 1 , 10, 100, 1000],
             'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
             'kernel' : ['rbf']}
```

```python
grid = GridSearchCV(SVC(),param_grid = param_grid, cv = 5, verbose = 3, refit = True )
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
ypred4 = grid.predict(X_test)
cr = classification_report(y_test, ypred4)
cm = confusion_matrix(y_test, ypred4)
print(cr)
print(cm)
```
